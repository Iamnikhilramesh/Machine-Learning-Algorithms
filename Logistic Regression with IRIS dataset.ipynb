```python
# import models 
from sklearn.linear_model import LogisticRegression
```

```python
# upload data to data variable as aa data frame
data = pd.read_csv('Iris.csv')
```

```python
#prepare data with x and y variable to fit into the model
x = data.drop(['Id','Species'],axis=1)
y = data['Species']
```

```python
#Prepare the Logistic Regression model 
log_reg = LogisticRegression()
```

```python
# fit and predict the values without splitting the dataset into train and test
log_reg.fit(x,y)
y_predict = log_reg.predict(x)
print(metrics.accuracy_score(y,y_predict))
```

```python
#split the data set to train and test model
x_test,x_train,y_test,y_train = train_test_split(x,y,test_size=0.4)
```

```python
# fit and predict the values with splitting the dataset into train and test
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
ypredict = logreg.predict(x_test)
print(metrics.accuracy_score(y_test,ypredict))
```
